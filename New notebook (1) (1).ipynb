```
import itertools as itl


class Ring:
    def __init__(self, val):
        self.__val = val
        self.next = None
        self.back = None

    def __str__(self):
        return f'{self.__val}'

    def data(self):
        return self.__val

    def rewrit(self, value, password=None):
        if password is None: password = input('Введите пароль: ')
        if password == 1:
            self.__val = value
        else:
            return 'Неверный пароль!'

    def Next(self, next):
        self.next = next
        return self

    def Back(self, back):
        self.back = back
        return self


class listrl:
    def __init__(self, val=None):
        if val is not None:
            try:
                iterator = iter(val)
            except TypeError:
                self.head = Ring(val)
                self.last = self.head
            else:
                self.head = Ring(val[0])
                self.last = self.head
                for i in val[1:]:
                    self.append(i)
                lrl = self.listrl()
                for i in lrl:
                    if i.next != None:
                        i.next.Back(i)
        else:
            self.head = None
            self.last = None

    def __str__(self):
        return "Listrl:" + str(list(self.show()))

    def append(self, val):
        if self.head == None:
            self.head = Ring(val)
            self.last = self.head
        else:
            self.last.Next(Ring(val))
            self.last = self.last.next.Back(self.last)

    def Head(self, var):
        if self.head == None:
            self.head = Ring(var)
            self.last = self.head
        else:
            self.head = Ring(var).Next(self.head)
            self.head.next.Back(self.head)

    def show(self):
        if self.head == None: return None
        a = self.head
        while a.next != None:
            yield a.data()
            a = a.next
        else:
            yield a.data()

    def showreverse(self):
        if self.head == None: return None
        a = self.last
        while a.back != None:
            yield a.data()
            a = a.back
        else:
            yield a.data()

    def listrl(self):
        if self.head == None: return None
        a = self.head
        while a.next != None:
            yield a
            a = a.next
        else:
            yield a

    def rlistrl(self):
        if self.head == None: return None
        a = self.last
        while a.back != None:
            yield a
            a = a.back
        else:
            yield a

    def is_empty(self):
        if self.head == None:
            return True
        else:
            return False

    def find(self, number=None, value=None):
        if number != None:
            if number > self.len() - 1 or number < 0: return 'IndexError'
            a = self.head
            for i in range(number):
                a = a.next
            return a

    def rewrit(self, number, value):
        self.find(number).rewrit(value, 1)

    def __len__(self):
        return self.len()

    def len(self):
        if self.head == None:
            return 0
        else:
            k = 0
        a = self.head
        while a.next != None:
            a = a.next
            k += 1
        else:
            k += 1
        return k

    # develop
    def pop(self, number=None):
        fl = False
        if number is None or number == self.len() - 1: number = self.len() - 1; fl = True
        if self.head is None: return None
        if number > self.len() - 1 or number < 0: return 'IndexError'

        if number == 0:
            var = self.head.data()
            if self.head.next is not None:
                if self.head.next != self.last:
                    self.head = self.head.next
                    self.head.Back(None)
                else:
                    self.head = self.head.next
                    self.head.Back(None)
            else:
                self.head = None
                self.last = None
            return var

        ring = self.find(number - 1)
        var = ring.next.data()
        if fl == True:
            ring.next = None
            self.last = ring
        else:
            ring.Next(ring.next.next)
            ring.next.Back(ring)
        return var


l = listrl([1, 2, 3, 4])
l.append(5)
l.Head(0)
print(list(l.show()))
for i in l.listrl():
    print(i.back, i.data(), i.next)
print(l)
l.pop(0)

l.pop(len(l) - 1)

l.pop(len(l) - 2)

l.pop(1)

print(list(l.showreverse()))

for i in l.listrl():
    print(i.back, i.data(), i.next)
```

```
class Queue(listrl):
    def __init__(self, val=None):
        super().__init__(val)
    def __str__(self):
        ql = super().__str__()
        return "Qlist" + ql[5:]
    def qappend(self, val):
        super().append(val)
    def Exit(self):
        return super().pop(0)
q = Queue([0,1,2,3,4,5])
print(q)
q.qappend(6)
print(q.Exit())
print(q)
```
